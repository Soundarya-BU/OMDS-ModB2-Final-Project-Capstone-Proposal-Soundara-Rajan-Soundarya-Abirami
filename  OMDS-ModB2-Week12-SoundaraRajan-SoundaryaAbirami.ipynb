# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

1. Side-by-side plots 

Advantages:

Easy to focus on each week's trend individually.

Avoids clutter, especially if lines overlap.

Allows for separate styling/scaling if needed.

Disadvantages:

Harder to directly compare values across weeks (you need to shift your eyes back and forth).

Axis labels might not be aligned or scaled consistently unless manually adjusted.

Less efficient in terms of space.

2. Single plot with both time series overlaid

Advantages:

Easier to compare patterns between the two weeks directly (e.g., is Tuesday busier in Week 2?)

More compact, space-efficient

Can quickly identify shifts, overlaps, or outliers

Disadvantages:

Can get visually cluttered, especially if lines are too close or too similar in color

Without clear legend or color distinction, viewers may get confused

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Department': ['HR', 'IT', 'Sales', 'Marketing', 'Finance', 'Operations'],
    'Positive': [60, 55, 45, 50, 65, 40],
    'Negative': [-20, -25, -35, -30, -15, -40]
}

df = pd.DataFrame(data)
df = df.set_index('Department')

# Plot setup
plt.figure(figsize=(10, 6))
plt.barh(df.index, df['Positive'], color='#4CAF50')
plt.barh(df.index, df['Negative'], color='#F44336')

# Styling
plt.axvline(0, color='black', linewidth=0.8)
plt.title('Employee Sentiment by Department')
plt.xlabel('Percentage of Responses')
plt.xticks(range(-50, 81, 10))
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Annotations (optional)
for i in df.index:
    plt.text(df.loc[i, 'Positive'] + 1, i, f"{df.loc[i, 'Positive']}%", va='center', color='black')
    plt.text(df.loc[i, 'Negative'] - 6, i, f"{abs(df.loc[i, 'Negative'])}%", va='center', color='black')

plt.show()
```
